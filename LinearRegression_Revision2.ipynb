```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_excel("AAPL.xlsx")
```

```python
df
```

```python
x=df.iloc[ :, :-1].values
x
```

```python
y=df.iloc[ :, -1].values
y
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x,y)
```

```python
plt.scatter(x,y, color="red")
plt.plot(x, regressor.predict(x), color="blue")
plt.show()
```
