```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_excel("AAPL.xlsx")
df
```

```python
x=df.iloc[ :, :-1].values
x
```

```python
y=df.iloc[ :, -1].values
y
```

```python
#It splits the data into two, train and test. It tries to predict using train data and checks the accuracy using test data.
#Here, it is divided in the ratio 8:2(test_size=0.2)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
#Fitting the train data to the variable "regressor"
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#Predicting data for the period "x_test" using past data(in regressor) and storing it to y_pred
y_pred = regressor.predict(x_test)
```

```python
plt.scatter(x_test, y_test, color=("red"))
plt.scatter(x_train, y_train, color=("green"))
plt.plot(x_test, y_pred, color="blue") #The predicted data
plt.show()
```
